It is assumed that the mean systolic blood pressure is μ = 120 mm Hg. In the Honolulu Heart Study, a sample of n = 100 people had an average systolic blood pressure of 130.1 mm Hg with a standard deviation of 21.21 mm Hg. Is the group significantly different (with respect to systolic blood pressure!) from the regular population?

### Steps 1 and 2: setting the hypotheses (null and alternate)

The null hypothesis that would need to be reviewed is:
"there is no significant difference between the mean systolic blood pressure of the Honolulu Heart Study sample and the population, and any observed difference is due to random sampling variability."

Symbolically, the null hypothesis can be expressed as:

H0: μ = 120 mm Hg,

where μ represents the population mean systolic blood pressure.

On the other hand, the alternate hypothesis is the opposite of the null hypothesis and represents the possibility of a significant difference between the Honolulu Heart Study sample mean and the population mean systolic blood pressure.

In this case, the alternate hypothesis would be:

H1: μ ≠ 120 mm Hg

This means that there is a significant difference between the sample mean and the population mean systolic blood pressure, and the difference is not likely due to random sampling variability.

### Step 3: determining if it is a one-tailed or a two-tailed test

This is a two-tailed test.

In a two-tailed test, the alternative hypothesis suggests that there is a significant difference between the sample mean and the assumed population mean in either direction, either higher or lower than the population mean.

In this case, the alternative hypothesis is Ha: μ ≠ 120 mm Hg, which means that the sample mean could be significantly different from the population mean either in the positive or negative direction. Therefore, this is a two-tailed test.

### Step 4: deciding a statistics test based on the information available

The population standard deviation is known in the given exercise, and the sample size (n = 100) is relatively large.

Therefore, we can use a z-test to test the null hypothesis that the mean systolic blood pressure in the Honolulu Heart Study sample is not significantly different from the assumed population mean of 120 mm Hg.

The z-test is an appropriate statistical test to use when the population standard deviation is known, and the sample size is sufficiently large. In this case, since the population standard deviation is known, we can use a z-test to determine the probability that the observed difference between the sample mean and the assumed population mean is due to chance or represents a true difference in means.

The test statistic for a z-test is calculated as:

z = (sample_mean - pop_mean) / (pop_std/math.sqrt(n))

### Step 5: level of significance and critical value (Z Value)

In this case, a significance level of 0.05 would be appropriate, which means that we are willing to accept a 5% chance of making an error.

To calculate the critical value for a two-tailed z-test (Z Value), we will follow the steps below: 

Finding the z critical value if the significance level is 0.05:

Step 1:

Divide the significance level α by 2.

α/2 = 0.05/2

α/2 = 0.025

Step 2:

Subtract α/2 from 1.

1 - α/2 = 1 – 0.025

1 - α/2 = 0.975

#### Z critical value = ±1.96 for this two-tailed test.

If the calculated test statistic falls between -1.96 and +1.96, we will fail to reject the null hypothesis because the test statistic would not be  extreme enough to provide evidence against the null hypothesis.

"Failing to reject" the null hypothesis would mean that we do not have enough evidence to reject it in favor of the alternative hypothesis. This is not the same as accepting the null hypothesis as absolutely true. Instead, we would simply not have enough evidence to conclude that the alternative hypothesis is true, so we would continue to assume that the null hypothesis is true.

If the calculated test statistic falls outside the range of -1.96 to +1.96, then we would reject the null hypothesis in favor of the alternative hypothesis. This would mean that we would conclude that the sample of 100 people is significantly different from the population mean of 120 mm Hg with a certain level of confidence (in this case, 95%).

### Step 6: calculations

```python
import math

sample_mean = 130.1
pop_mean = 120
pop_std = 21.21
n = 100

z = (sample_mean - pop_mean)/(pop_std/math.sqrt(n))
z
```

### Step 7: conclusions

The above calculation means that the sample mean is 4.76 standard errors away from the assumed population mean under the null hypothesis.

In other words, the sample mean is significantly different from the population mean assumed under the null hypothesis.

Therefore, we reject the null hypothesis and conclude that the group in the Honolulu Heart Study is significantly different from the regular population with respect to systolic blood pressure.

If we reject the null hypothesis, we can say that the difference between the sample mean and population mean did not occur by chance, at least not to the extent that the null hypothesis assumed. Rejecting the null hypothesis means that the observed difference is statistically significant, and it is unlikely that this result happened by chance alone.
